```python
# 내 데이터
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

data_path = "./datasets/numbers/data.yaml" 
model.train(data=data_path, epochs=30, imgsz=640)
```

```python
# 팀원 데이터
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

data_path = "./datasets/num/data.yaml" 
model.train(data=data_path, epochs=30, imgsz=640)
```

```python
# 하이퍼파라미터튜닝
from ultralytics import YOLO
model = YOLO('runs/detect/train17/weights/best.pt')
model.tune(
    data='./datasets/num/data.yaml',
    epochs=30,
    imgsz=640,
    iterations=10,
    val=True
)
```

```python
from ultralytics import YOLO
import yaml

data_path = "./datasets/num/data.yaml"
with open(r"runs\detect\tune4\best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)
model.train(data=data_path, epochs=30, imgsz=640, batch=20, **hyp_dict)
```
